```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
arr= np.linspace(-100,100,300)
```

```python
arr2= [sigmoid(i) for i in arr]
```

```python
import matplotlib.pyplot as plt

plt.plot(arr2)
```

```python
arr= np.linspace(-1000,1000,1000)
arr2= [sigmoid(i) for i in arr]
plt.plot(arr2)
```

```python
import seaborn as sns
```

```python
arr= np.linspace(-10,10,100)
arr2= [sigmoid(i) for i in arr]
plt.plot(arr2, label='Sigmoid')
plt.axhline(y=np.array(arr2).mean(), color='r', linestyle='--', label='Threshold')
plt.legend()
```
